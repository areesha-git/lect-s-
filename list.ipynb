```python
# W.A.P which will take input from the user and increase even no by 5 and decrease by 5
lst = eval(input("enter list"))
print(lst)
for i in range(len(lst)):
    if lst[i]%2==0: lst[i]+=5
    else:lst[i]-=5
print(lst)
```

```python
# W.A.P find the largest elements and 2nd largest element

lst=eval(input("Enter the list"))
print(lst)
largest=lst[0]
seclargest=lst[0]
for i in range(0,len(lst)):
    if lst[i]>largest:
        largest=lst[i]
for i in range(0,len(lst)):
    if lst[i]>seclargest and lst[i]!=largest:
        seclargest=lst[i]
print("Largest element is:", largest)
print("Second largest element is:", seclargest)
```
