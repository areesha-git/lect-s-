```python
# W.A.P to print first and last elements of each list of list

lst = [[23,45,76],[45,3,12],[3,5,12]]
lst1 =[]
for i in lst :
    lst1.append(i[0])
    lst1.append(i[-1])
print(lst1)
```

```python
# W.A.P to calculate and print the sum of the max element of each list of lists
lst = [[1,2,4,5],[3,5,4,3],[4,5,3,2]]
total = 0 
for sublist in lst:
    total+=max(sublist)
print(total)    
```

```python
# sum of the element
lst = [[1,2,4,5],[3,5,4,3],[4,5,3,2]]
total = 0 
for sublist in lst:
    total+=sum(sublist)
print(total)   
```

```python
# W.A.P to divide a list into a equal parts
lst = [2,4,6,8,3,5,8,9]
print(sum(lst[:len(lst)//2]),sum(lst[len(lst)//2:]))
```
